# Probability Distribution Functions in Python

```python
from numpy import random

x = random.choice([3, 5, 7, 9], p=[0.1, 0.3, 0.5, 0.1], size=(100))

print(x)
```

```python
x = random.choice([3, 5, 7, 9], p=[0.1, 0.3, 0.6, 0.0], size=(3, 5))

print(x)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
x = random.choice([3, 5, 7, 9], p=[0.1, 0.3, 0.5, 0.1], size=(100))
sns.displot(x)
plt.show()
```

# Normal Distribution

```python
x = random.normal(size=100)

print(x)
```

```python
x = random.normal(loc=1, scale=2, size=(2, 3))

print(x)
```

```python
sns.displot(random.normal(size=1000), kde=True)
plt.show()
```

# Binomial  Distribution

```python
x = random.binomial(n=10, p=0.5, size=30)

print(x)
```

```python
import seaborn as sns

sns.displot(random.binomial(n=10, p=0.5, size=1000), kde=False)

plt.show()
```

# Poisson Distribution

```python
sns.displot(random.poisson(lam=2, size=1000), kde=False)

plt.show()
```

# Uniform Distribution

```python
random.uniform(size=10)
```

```python
sns.displot(random.uniform(size=100000), kde=False)

plt.show()
```

# Logistic Distribution

```python
sns.displot(random.logistic(size=1000),  kde=True)

plt.show()
```

# Exponential Distribution

```python
sns.displot(random.exponential(size=1000),  kde=True)

plt.show()
```

# Chi Square Distribution

```python
sns.displot(random.chisquare(df=1, size=1000),  kde=True)

plt.show()
```
